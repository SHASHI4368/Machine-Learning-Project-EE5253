## PCA

```python
# Selecting the feature dataframe
x = df.drop('Payment Method', axis=1)
```

```python
from sklearn.decomposition import PCA
```

```python
# Finding the number of components to retain 80% + variance
pca = PCA(n_components=10)
df_pca = pca.fit_transform(x)
component_df = pd.DataFrame(data=df_pca)
np.cumsum(pca.explained_variance_ratio_)
```

```python
# Selecting the feature dataframe using PCA
x = component_df
x.head()
```

## KNN Modelling

## Initializing the features and the predictor variables

```python
# Selecting the target dataframe
y = df['Payment Method']
```

## Splitting the dataset

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.25, random_state=101)
```

### Importing the KNN Libraries

```python
from sklearn.neighbors import KNeighborsClassifier
```

### Applying hyperparameter optimization

```python
from sklearn.model_selection import GridSearchCV
```

```python
knn_model = KNeighborsClassifier()
```

```python
knn_params = {'n_neighbors': np.arange(1, 50),
              'metric': ['minkowski', 'euclidean', 'manhattan'],
              'weights': ['uniform', 'distance']}
```

```python
knn_cv_model = GridSearchCV(estimator=knn_model, param_grid=knn_params, cv=10)
```

```python
knn_cv_model.fit(x_train, y_train)
```

```python
knn_cv_model.best_params_
```

```python
knn_tuned = KNeighborsClassifier(**knn_cv_model.best_params_)
knn_tuned.fit(x_train, y_train)
```

```python
y_pred = knn_tuned.predict(x_test)
```

### Validation

```python
import sklearn.metrics as metrics
print('Accuracy:', metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
```

### Using cross validation

```python
cv_scores = cross_val_score(estimator=knn_tuned, X=x_train, y=y_train, cv=10)
print('Mean Accuracy: ', cv_scores.mean())
```
